```python
# Run this function if entity in CREATIVE, ADSET, CAMPAIGN
def percentage(above, total,entity):
    percent = above / total * 100
    print(f"The percentage above spend for {entity} is :",percent,"%")
    
a = list(data['sum'])
a = [float(i) for i in a]
total = sum(a)
print("The total spend is :", total)

# filtering data
data1 = data.query('goal == "above"')

a = list(data1['sum'])
a = [float(i) for i in a]
above = sum(a)
print("The above spend is :",above)


percentage(above,total,entity)
```
