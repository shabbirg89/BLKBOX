```python
def adds_folder(ad_account_ids):  
  for each_add_id in ad_account_ids: 
    #ad_acc = input("Enter the Adaccount_id of your choice without quotes:")
    ad_acc = each_add_id
    print("The number of images in {a} ad_account is :".format(a=each_add_id))

    path = "/content/drive/MyDrive/BLKBOX/BLKBOX_CREATIVES/{a}".format(a = ad_acc)
    #os.mkdir(path)

    if not os.path.exists(path):
        os.makedirs(path)

    conn = psycopg2.connect(user = "ds_readonly", password = "blkbox2020!",host = "db.blkbox.ai",port = "5432",database = "blkbox")
    sql = "select adaccount_id,asset_id,CAST(asset_metadata as jsonb)->>'thumbnail_url' as thumbnail_url,performance from public.fb_app_assets where adaccount_id = '{a}';".format(a = ad_acc)
    data = pd.read_sql_query(sql, conn)
    conn = None
  # data.head()
    data['thumbnail_url'].replace('', np.nan, inplace=True)
  # data.head()
  # data.tail()
  # data.shape
    data.fillna(0, inplace = True)
    data.isnull().sum()

    j = 0 
    k = 0

    urls = data[['thumbnail_url']] #save the url list as a dataframe
    rows = []

    ids = data[['adaccount_id']] #save the url list as a dataframe
    rows1 = []

    asset_id = data[['asset_id']] #save the url list as a dataframe
    rows2 = []

    for index, i in urls.iterrows():
      rows.append(i[-1])

    for index, j in ids.iterrows():
      rows1.append(j[-1])

    for index, k in asset_id.iterrows():
      rows2.append(k[-1])

      counter = 0

    save_path = '/content/drive/MyDrive/BLKBOX/BLKBOX_CREATIVES/{a}'.format(a=ad_acc)
    data.tail()

  # print(rows)
    rows = filter(lambda x: x != 0, rows)

    #for i in rows:
    for i, j ,k in zip(rows, range(len(rows1)),range(len(rows2))):
        file_name = rows1[j] + '-' + str(rows2[k]) + ':' + str(counter) + '.png'
        completeName = os.path.join(save_path,file_name)
        response = requests.get(i)
     #  print(file_name)
     #  print(completeName)
        file = open(completeName, "wb")
        file.write(response.content)
        file.close()
        counter += 1
        j += 1
        k += 1
    print(counter)
```
