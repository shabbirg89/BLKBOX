```python
import requests
import time
```

```python
def download(url,name):
  r = requests.get(url, allow_redirects=True)
  open('/content/drive/MyDrive/BLKBOX_VIDEO/video/{a}.mp4'.format(a=name), 'wb').write(r.content)
  print("Download_Complete :",name)
```
